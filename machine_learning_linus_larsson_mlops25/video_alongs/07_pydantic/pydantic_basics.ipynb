## validation with pydantic

```python
class Person:
    def __init__(self, name: str, gender: str, age: int) -> None:
        if not isinstance(name, str):
            raise TypeError(f"name must be of type str not {type(name)}")
        self.name = name
        self.gender = gender
        self.age = age
        
try:
    person_1 = Person(name = 3, age = 22, gender = "male")
except TypeError as e:
    print(e)
```

```python
class Person:
    def __init__(self, name: str, gender: str, age: int) -> None:
        if not isinstance(name, str):
            raise TypeError(f"name must be of type str not {type(name)}")
        self.name = name
        self.gender = gender

        if not isinstance (age, int):
            raise TypeError(f"age must be of the type int not {type(age)}")
        if not 0 <= age < 125:
            raise ValueError(f"age must be between 0 - 124 not {age}")
        self.age = age
    

    def __repr__(self):
        return f"Person({self.name}, {self.gender}, {self.age})"

try:
    person_3 = Person(name = "Gun", age = -45, gender = "female")
except ValueError as e:
    print(e)
```

```python
person_4 = Person(name = "Britt", age = 60, gender = "female")
person_4
```

```python
class Person:
    def __init__(self, name: str, gender: str, age: int) -> None:
        if not isinstance(name, str):
            raise TypeError(f"name must be of type str not {type(name)}")
        self.name = name
        self.gender = gender
        self.age = age

    @property
    def age(self):
        return self._age
    
    @age.setter
    def age(self, age):
        if not isinstance (age, int):
            raise TypeError(f"age must be of the type int not {type(age)}")
        if not 0 <= age < 125:
            raise ValueError(f"age must be between 0 - 124 not {age}")
        self._age = age
    

    def __repr__(self):
        return f"Person({self.name}, {self.gender}, {self.age})"

person_5 = Person(name = "Maj", age = 60, gender = "female")
person_5
```

```python
try:
    person_5.age = 125
except ValueError as e:
    print(e)
```

```python
person_6 = Person(name = "Gunhilda", age = 60, gender = "female")

try:
    person_5.age = "10"
except ValueError as e:
    print(e)
```

## Validate using pydantic

```python
from pydantic import BaseModel

class Person(BaseModel):
    name: str
    gender: str
    age: int

person_6 = Person(name = "kristina", gender = "female", age = 45)
person_6
```

```python
person_6.age = 36
person_6
```

```python
# Issue
person_6.age = "thirty"
person_6
```

```python
from pydantic import ValidationError
try:
    Person(name = 4533, gender = "female", age = 45)
except ValidationError as e:
    print(e)
```

```python
# str "45" is correctred into 45
person_7 = Person(name = "Harryiet", gender = "female", age = 45)
```

```python
from pydantic import ConfigDict
class Person(BaseModel):
    name: str
    gender: str
    age: int

    model_config = ConfigDict(validate_assignment=True)
try:
    Person(name = 4533, gender = "female", age = 45)
except ValidationError as e:
    print(e)

person_6.age = 10
person_6
```

```python
try:
    Person(name = "klas", gender = "male", age = 4)
except ValidationError as e:
    print(e)
```

## Add age validation

```python
from pydantic import BaseModel, Field, ValidationError
from pydantic import ConfigDict
from typing import Literal

class Person(BaseModel):
    name: str
    gender: Literal["Male", "Female"]
    age: int = Field(gt = -1, lt = 125)

    model_config = ConfigDict(validate_assignment=True)


try:
    Person(name = "klas", gender = "male", age = -4)
except ValidationError as e:
    print(e)
```
